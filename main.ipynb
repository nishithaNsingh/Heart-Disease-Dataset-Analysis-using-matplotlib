```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Dataset Heart Disease.csv')
```

```python
data.tail(10)
```

```python
plt.scatter(data['age'],data['max heart rate'])
plt.show()
```

```python
# Histogram of Age
plt.figure(figsize=(10, 6))
plt.hist(data['age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot comparing cholesterol levels between patients with and without heart disease
plt.figure(figsize=(8, 6))
data.boxplot(column='cholesterol', by='target', grid=False)
plt.title('Cholesterol Levels by Heart Disease Status')
plt.xlabel('Heart Disease (0: No, 1: Yes)')
plt.ylabel('Cholesterol Level')
plt.show()
```

```python
#Bar plot showing the count of each chest pain type
plt.figure(figsize=(8, 6))
data['chest pain type'].value_counts().sort_index().plot(kind='bar', color='red')
plt.title('Count of Chest Pain Types')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Scatter plot of age vs. maximum heart rate
plt.figure(figsize=(8, 6))
plt.scatter(data['age'], data['max heart rate'], color='green')
plt.title('Age vs. Maximum Heart Rate')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate')
plt.show()
```

```python
# Pie chart showing the distribution of heart disease cases
plt.figure(figsize=(8, 6))
heart_disease_counts = data['target'].value_counts()
labels = ['No Heart Disease', 'Heart Disease']
plt.pie(heart_disease_counts, labels=labels, autopct='%1.1f%%', colors=['skyblue', 'salmon'])
plt.title('Distribution of Heart Disease Cases')
plt.show()
```

```python
# Stacked bar plot showing the distribution of heart disease cases by gender
gender_heart_disease = data.groupby(['sex', 'target']).size().unstack()
gender_heart_disease.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'], figsize=(8, 6))
plt.title('Distribution of Heart Disease Cases by Gender')
plt.xlabel('Gender (0: Female, 1: Male)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(['No Heart Disease', 'Heart Disease'])
plt.show()
```

```python
# Bar plot showing the distribution of heart disease cases by chest pain type
plt.figure(figsize=(8, 6))
chest_pain_heart_disease = data.groupby(['chest pain type', 'target']).size().unstack()
chest_pain_heart_disease.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'], figsize=(8, 6))
plt.title('Distribution of Heart Disease Cases by Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(['No Heart Disease', 'Heart Disease'])
plt.show()
```

```python
# Violin plot of age distribution by heart disease status
plt.figure(figsize=(8, 6))
sns.violinplot(x='target', y='age', data=data, palette='Set2')
plt.title('Age Distribution by Heart Disease Status')
plt.xlabel('Heart Disease (0: No, 1: Yes)')
plt.ylabel('Age')
plt.show()
```

```python
# Bar plot showing the distribution of heart disease cases by fasting blood sugar levels
plt.figure(figsize=(8, 6))
blood_sugar_heart_disease = data.groupby(['fasting blood sugar', 'target']).size().unstack()
blood_sugar_heart_disease.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'], figsize=(8, 6))
plt.title('Distribution of Heart Disease Cases by Fasting Blood Sugar Level')
plt.xlabel('Fasting Blood Sugar (0: < 120 mg/dl, 1: >= 120 mg/dl)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(['No Heart Disease', 'Heart Disease'])
plt.show()
```
